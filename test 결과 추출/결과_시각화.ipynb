```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
%matplotlib inline

def read_file(path):
  df = pd.read_csv(path, delimiter = '\t')
  return df

def returnWhichBig_type(a, b, c):
  result = []
  for i in range(0, len(a)):
    if a[i]>b[i] and a[i]>c[i]:
      result.append('interviewReview')
    elif b[i] > c[i] :
      result.append('failureReview')
    else:
      result.append('recruitReview')
  return result
```

```python
df = read_file('/content/drive/My Drive/닥터송 여름 프로젝트/4. 대-스타 해결 2/데이터/ResultData/0930 결과/리뷰타입_results.tsv')

df['type_label'] = returnWhichBig_type(df['interviewReview'].values.tolist(), df['failureReview'].values.tolist(), df['recruitReview'].values.tolist())

#col : dataset 기준으로 따로, col_wrap : n개씩 나눠서 보기, ci : 데이터 분포에 관한 신뢰구간, scatter_kws : 점 크기, 투명도

sb.lmplot(x='reviewIndex', y='recruitReview', data=df, height=5, fit_reg=False, scatter_kws={"s":10, "alpha":1}, hue='type_label', ci=10)
sb.lmplot(x='reviewIndex', y='interviewReview', data=df, height=5, fit_reg=False, scatter_kws={"s":10, "alpha":1}, hue='type_label', ci=10)
sb.lmplot(x='reviewIndex', y='failureReview0', data=df, height=5, fit_reg=False, scatter_kws={"s":10, "alpha":1}, hue='type_label', ci=10)


# plt.title('modeling result')
```
